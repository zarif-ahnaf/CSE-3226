**Problem Statement**: Take a 5x6 NumPy array and plot the histogram of the array with proper labels (without built in histogram function). Show both histograms side by side in a subplot with a title. Show input NumPy array into an image.

```python
import numpy as np

import matplotlib.pyplot as plt

array = np.array([
    [255,0,1,100,25,101],
    [10,200,70,80,120,150],
    [95,30,30,81,96,77],
    [87,89,220,250,100,10],
    [18,7,221,21,8,15]
])

hist = np.zeros(256)

for i in range(array.shape[0]):
    for j in range(array.shape[1]):
        hist[array[i,j]]+=1

hist2,bins = np.histogram(array,256,[0,256])

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(hist)

plt.subplot(1,3,2)
plt.title("Histogram (NumPy)")
plt.plot(hist2)

plt.subplot(1,3,3)
plt.title("Image Representation")
plt.imshow(array, cmap='gray')
plt.colorbar()

plt.show()
```
