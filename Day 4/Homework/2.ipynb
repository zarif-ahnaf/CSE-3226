**Problem Statement**: Take an image and plot the histogram of the array with proper labels (without built in histogram function). Show both histograms side by side in a subplot with a title.

```python
import cv2

import numpy as np

import matplotlib.pyplot as plt

array = cv2.imread("./a.jpg")
hist = np.zeros(256)

for i in range(array.shape[0]):
    for j in range(array.shape[1]):
        hist[array[i,j]]+=1

hist2,bins = np.histogram(array,256,[0,256])

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(hist)

plt.subplot(1,3,2)
plt.title("Histogram (NumPy)")
plt.plot(hist2)

plt.subplot(1,3,3)
plt.title("Image Representation")
plt.imshow(array, cmap='gray')
plt.colorbar()
plt.show()
```
