# Take 3 lists and print their sum

```python
list1 = [1,2,3]
list2 = [4,5,6]
list3 = [7,8,9]

print(f"List 1, {list1}")
print(f"List 2, {list2}")
print(f"List 3, {list3}")

for i in [list1,list2,list3]:
    _sum = 0
    for j in i:
        _sum+=j
    print(_sum)

print(f"Sum of list 1: {sum(list1)}")
print(f"Sum of list 2: {sum(list2)}")
print(f"Sum of list 3: {sum(list3)}")

print(f"Sum of three Lists: {sum(list1+list2+list3)}")
```
