```python
# 90 degree (Clockwise)
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load and grayscale
img = cv2.imread("./a.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
height, width = img_gray.shape

# Create a new empty array with SWAPPED dimensions (width, height)
img_rotated = np.zeros((width, height))

for i in range(height):
    for j in range(width):
        img_rotated[j][height - 1 - i] = img_gray[i][j]

# Displaying results
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap="gray")
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(img_rotated, cmap='gray')
plt.title("Rotated 90° CW")
plt.show()
```

```python
# -90 Degree Rotation (Counter Clockwise)
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "./a.jpg"
img = cv2.imread(image_path)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
height, width = img_gray.shape

img_rotated_neg90 = np.zeros((width, height))

for i in range(height):
    for j in range(width):
        img_rotated_neg90[width - 1 - j][i] = img_gray[i][j]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap="gray")
plt.title(f"Original Image\n({height}x{width})")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_rotated_neg90, cmap='gray')
plt.title(f"Rotated -90° (CCW)\n({width}x{height})")
plt.axis('off')

plt.tight_layout()
plt.show()
```
