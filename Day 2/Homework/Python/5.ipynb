**Problem Statement**:

Take an 8-bit grayscale image and apply Thresholding such that the pixel having intensity values below 150 will be replaced by 0 and pixels having intensity value equal or above 150 will be replaced by 255. Display input and output image side by side in a subplot with a title.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image_path = "./a.jpg"
img = cv2.imread(image_path)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
height,width = img_gray.shape
img_mirrored = np.zeros((height,width))
for i in range(height):
  for j in range(width):
    pixel =  img_gray[i][j]
    if (pixel <150):
      pixel = 0
    else:
      pixel = 255
    img_mirrored[i][j] = pixel
plt.subplot(1,2,1)
plt.imshow(img_gray,cmap="gray")
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(img_mirrored,cmap='gray')
plt.title("Intensity Value")
```
