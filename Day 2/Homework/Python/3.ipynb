**Problem Statement**:

Take a grayscale image and generate the mirror image of the original image. Show input and output image side by side in a subplot with a title.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./a.jpg')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

mirrored_img = img_gray.copy()
rows, cols = img_gray.shape
for i in range(rows):
    for j in range(cols):
        mirrored_img[i, j] = img_gray[i, cols - 1 - j]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original')

plt.subplot(1, 2, 2)
plt.imshow(mirrored_img, cmap='gray')
plt.title('Mirrored')

plt.show()
```
