```python
import numpy as np

original_arr = np.array([[255,0,1,100,25,101],[10,200,70,80,120,150],[95,30,30,81,96,77],[87,89,220,250,100,10],[18,7,221,21,8,15]])

rev = []

def flip(x):
    i, j = 0, len(x) - 1

    while i < j:
        x[i], x[j] = x[j], x[i]
        i += 1
        j -= 1
    return x
for i in original_arr:
  rev.append(flip(i))
rev
```
